```python
def draw_graph(graph, explored, node_positions=None, start=None, goal=None, path=None):
    """Visualize results of graph search"""

    labels = {}
    for node in graph:
        labels[node] = node

    if node_positions is None:
        node_positions = networkx.spring_layout(graph)

    nx.draw_networkx_nodes(graph, node_positions)
    nx.draw_networkx_edges(graph, node_positions, style='dashed')
    nx.draw_networkx_labels(graph, node_positions, labels)

    nx.draw_networkx_nodes(graph, node_positions, nodelist=explored, node_color='g')

    if path is not None:
        edges = [(path[i], path[i + 1]) for i in range(0, len(path) - 1)]
        nx.draw_networkx_edges(graph, node_positions, edgelist=edges, edge_color='g')

    if start is not None:
        nx.draw_networkx_nodes(graph, node_positions, nodelist=[start], node_color='b')

    if goal is not None:
        nx.draw_networkx_nodes(graph, node_positions, nodelist=[goal], node_color='y')

    #plt.plot()
    plt.show()
```

```python
graph.add_nodes_from(nodes)
```

```python
for node in graph.nodes:
    x, y = graph.node[node]['pos']
    neighbors = [(np.clip(x + del_x, 0, N-1), np.clip(y + del_y, 0, M-1)) for del_x, del_y in steps]
    edges = []
    for neighbor in neighbors:
        if neighbor != node:
            edge_weight = ((neighbor[1] - node[1])**2 + (neighbor[0] - node[0])**2)**0.5
            edges.append((node, neighbor, edge_weight))
    graph.add_weighted_edges_from(edges)
```
