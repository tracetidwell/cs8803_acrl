```python
def fitness(game):
    field = np.frombuffer(tetris.getByteArray(), dtype=np.int32).reshape(game.getRows(), game.getCols())
    top = np.array(list(game.getTop()))
    
    height = sum(top)
    
    lines = game.getRowsCleared()
    
    holes = 0
    for i in range(game.getCols()):
        holes += sum(field[:top[i], i] == 0)
        
    bumpiness = sum(abs(top[:-1] - top[1:]))
    
    return (5 * lines**2) - (0.25 * height) +  - (0.25 * holes) - (0.10 * bumpiness)   
```

```python
class Tetris(object):
    
    def __init__(self, tetris, fitness_fn=fitness):
        self.tetris = tetris
        self.curr_fitness = 0
        self.fitness_fn = fitness_fn
        
    def __getattr__(self, item):
        return getattr(self.tetris, item)
    
    def make_move(self, action):
        self.tetris.makeMove(action)
        new_fitness = self.fitness_fn(self.tetris)
        reward = new_fitness - self.curr_fitness
        self.curr_fitness = new_fitness
        return reward
```
